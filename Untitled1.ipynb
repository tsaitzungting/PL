```python
import requests 
import json
import time
from time import sleep
import pandas as pd 
```

```python
#  裁判家網址：https://www.lawplus.com.tw/
#  搜索內容:槍砲刀械
#  我想找罰金

hi_teacher=[]
for page in range(1,8):
    urlfirst = "https://www.lawplus.com.tw/rest/search/report?querySentence=%E6%A7%8D%E7%A0%B2%E5%88%80%E6%A2%B0&keyword=%E6%A7%8D%E7%A0%B2%E5%88%80%E6%A2%B0&prevKeyword=%E6%A7%8D%E7%A0%B2%E5%88%80%E6%A2%B0&date=&money=&sentence=&caseNum=&caseTypes=&courts=&levels=&jtypes=&tags=&issue=&main=&judge=&judgeTypes=&lawyer=&litigant=&prosecutor=&clerk=&rows=10&page=1&sortField=&_=1667413240806"
    urlsec = "&sortField=&_=1667413240806"
    link=urlfirst+str(page)+urlsec
    res = requests.get(link,verify=False)
    data=json.loads(res.text)
    temp_data=data['rows']

    for i in temp_data:
        hi_teacher.append(i['identifier'])
print(hi_teacher)
```

```python
#真正的網址
bitch=[] 
for r in hi_teacher:
    url="https://www.lawplus.com.tw/#reportDetail?id="+r
    bitch.append(url)
for i in bitch:
    print(i)  
```

```python
#爬蟲網址
bitchlink=[]
for r in hi_teacher:
    each="https://www.lawplus.com.tw/rest/search/report/"+r
    bitchlink.append(each)
for i in bitchlink:
    print(i)
```

```python
content=[]

for i in bitchlink:
    res=requests.get(i,verify=False)
    resquests=json.loads(res.text)
    
# dictionary存放

    content.append({
    '日期':resquests['response']['reportBase']['identifier'].split(',')[4], 
    '案號':resquests['response']['reportBase']['caseNum'],
    '內文':resquests['response']['reportBase']['content'],
    '法規':resquests['response']['report']['statute'],
    '法官' :resquests['response']['report']['presideJudge'],
    '裁判類型':resquests['response']['report']['type'],
    '案由':resquests['response']['reportBase']['issue']     
})
    sleep(1)
```
